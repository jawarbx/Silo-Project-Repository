```python
import pandas as pd

df = pd.read_csv("draft_data.csv")
df['LOC_SET'] = "////////"
df.head()
```

```python
org_set = set()
```

```python
test_idx = 0

test_ent = list(df.iloc[test_idx][1:])
test_ent
```

```python
import ast

for ent in test_ent:
    if type(ent) is str and ent != "////////":
        temp = ast.literal_eval(ent)
        if temp['entity_group'] == 'ORG':
            if temp['word'] != '.':
                org_set.add(temp['word']) 
org_set
```

```python
df['ORG_SET'][test_idx] = str(org_set)
```

```python
import ast
def parse_locs(ent_list):
    temp_org_set = set()
    for ent in ent_list:
        if type(ent) is str and ent != "////////":
            temp = ast.literal_eval(ent)
            if temp['entity_group'] == 'LOC':
                if temp['word'] != '.':
                    temp_org_set.add(temp['word']) 
    return str(temp_org_set)
```

```python
for i in range(4330):
    df['LOC_SET'][i] = parse_locs(list(df.iloc[i][1:]))
```

```python
desired = df[['Unnamed: 0','LOC_SET']]
```
