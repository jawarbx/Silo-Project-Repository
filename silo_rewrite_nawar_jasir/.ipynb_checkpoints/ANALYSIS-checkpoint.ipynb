```python
import pandas as pd
from scipy.stats import ttest_ind
df = pd.read_csv("final_data.csv").fillna("")
df.head()
```

```python
def grant_counter(grant_str):
    if grant_str == "":
        return 0
    else:
        return len((grant_str.strip('[]')).split("[]"))
```

```python
df['GRANT_COUNT'] = df['Grants_and_Funding'].apply(grant_counter)
```

# Analysis / Graphs

```python
#Parent Siloing in Whole Dataset
df['SILO'].value_counts(sort=False).plot(
    kind='bar',
    title="Frequency of Institutionally Siloed Literature (n=493)",
    xlabel="Siloed?",
    ylabel="Amount of Articles"
    )
```

```python
#Parent Siloing in Whole Dataset
df['DEPT_SILO'].value_counts(sort=False).plot(
    kind='bar',
    title="Frequency of Departmentally Siloed Literature (n=493)",
    xlabel="Siloed?",
    ylabel="Amount of Articles"
    )
```

```python
#Extreme Siloing in Whole Dataset
def extreme_siloing(df):
    if df['SILO'] == "Yes" and df['DEPT_SILO'] == "Yes":
        return "Yes"
    else:
        return "No"

df['EXTREME_SILO'] = df.apply(lambda x: extreme_siloing(x),axis=1)
df['EXTREME_SILO'].value_counts(sort=False).plot(
    kind='bar',
    title="Frequency of Extremely Siloed Literature (n=493)",
    xlabel="Siloed?",
    ylabel="Amount of Articles"
    )
```

```python
def box_plotter(continuous_variable, binary_variable, data):
    temp = data[[continuous_variable,binary_variable]]

    no_data = temp[temp[binary_variable] == "No"]
    yes_data = temp[temp[binary_variable] == "Yes"] 

    t_test = ttest_ind(no_data[continuous_variable], yes_data[continuous_variable])

    temp.boxplot(
        column=continuous_variable,
        by=binary_variable,  
    ).set_title(f"(p = {t_test[1]:.6f})").get_figure().suptitle(f"{continuous_variable} of {binary_variable} vs Non {binary_variable} Literature")
```
