```python
import pandas as pd
```

```python
df_full = pd.read_csv("full.csv")
df_full.head()
```

```python
df_full.fillna('', inplace=True)
df_full.iloc[10]
```

```python
affs = df_full['Affiliations']
affs.head()
```

```python
import json
import re
affs.fillna("")
list_of_list_of_affs = []
for aff in affs:
    list_of_list_of_affs.append(re.split(r'////|; ', aff))

result = list(set().union(*list_of_list_of_affs))
result
```

```python
len(result)
```

```python
#Finding Departments
import re
import regex


dept_reg = re.compile(r'depart*ment|division', flags= re.IGNORECASE)
dept = [k for k in result if dept_reg.findall(k) is not []]
dept
```

```python
#Find Universities
uni_reg = re.compile(r'universit*|college', flags= re.IGNORECASE)
france = re.compile(r'CHU|NHC', flags=re.IGNORECASE)

universities = {k for k in result if uni_reg.search(k) or france.search(k)}
len(universities)
```

```python
uni_list_names = pd.read_csv("world-universities.csv")
uni_name_set = set(uni_list_names['University of Andorra'])

for aff in result:
    for key in uni_name_set:
        if key.lower() in aff.lower():
            universities.add(aff)

len(universities)
```

```python
no_unis = list(set(result) ^ set(universities))
no_unis
```
